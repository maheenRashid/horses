```python
import util;
import visualize;
import os;
import preprocessing_data
import scipy.misc;
import cv2;
import numpy as np;
def reviseLimits(min_v,max_v,lim_v,aspect):
    length=max_v-min_v;
    new_length=length*aspect;
    buff=(new_length-length)/2;
    min_v_new=min_v-buff;
    max_v_new=max_v+buff;
    
    
    pad_bef=0;
    pad_aft=0;

    if min_v_new<0:
        pad_bef=abs(min_v_new);
        min_v_new=0;

    if max_v_new>lim_v:
        pad_aft =max_v_new-lim_v
        max_v_new=lim_v+pad_aft+pad_bef;


    return min_v_new,max_v_new,pad_bef,pad_aft;
    

def resizeIm(im_file,bbox,resize_size,out_file,aspect,vertical):
    im = cv2.imread(im_file, 1)
    rows, cols = im.shape[:2]
    
    if vertical:
        bbox[0],bbox[1],pad_bef,pad_aft=reviseLimits(bbox[0],bbox[1],cols,aspect);
        im=np.pad(im,((0,0),(pad_bef,pad_aft),(0,0)),'edge');
    else:
        bbox[2],bbox[3],pad_bef,pad_aft=reviseLimits(bbox[2],bbox[3],rows,aspect)
        im=np.pad(im,((pad_bef,pad_aft),(0,0),(0,0)),'edge');
        
    im=im[bbox[2]:bbox[3],bbox[0]:bbox[1],:];
    im = cv2.resize(im, (resize_size[1],resize_size[0]))
    cv2.imwrite(out_file,im);

# from IPython.display import Image

val_list_file='/home/laoreja/deep-landmark-master/dataset/train/testImageList.txt';
path_pre='/home/laoreja/deep-landmark-master/dataset/train';
face=True
out_dir='/home/SSD3/maheen-data/temp/distorted_humans'
util.mkdir(out_dir)
dir_server='/home/SSD3/maheen-data'
dummy_file='/home/SSD3/maheen-data/horse_project/data_check/face/npy/lfw_5590/Javier_Saviola_0001.npy'

resize_size=[224,224]
val_list=util.readLinesFromFile(val_list_file);
val_list=val_list[:100]
aspect=4;

path_im,bbox,anno_points = preprocessing_data.parseAnnoFile(val_list_file,path_pre,face);
im_files_all=[];
param_tuples=[(os.path.join(out_dir,'face_vert.txt'),'_vert.jpg',aspect,True),\
              (os.path.join(out_dir,'face_horz.txt'),'_horz.jpg',aspect,False),\
              (os.path.join(out_dir,'face_org.txt'),'_org.jpg',1,False)]
for out_file,post_curr,aspect_curr,vert_bool_curr in param_tuples:
    im_files_all=[];
    for idx,(im_path,bbox_curr,anno_points_curr) in enumerate(zip(path_im,bbox,anno_points)):
        if idx%100==0:
            print idx;
        out_file_curr=os.path.join(out_dir,str(idx)+post_curr);
        resizeIm(im_path,bbox_curr[:],resize_size,out_file_curr,aspect_curr,vert_bool_curr);
        im_files_all.append(out_file_curr);
        
    im_files_all=[line_curr+' '+dummy_file for line_curr in im_files_all]
    util.writeFile(out_file,im_files_all);

visualize.writeHTMLForFolder(out_dir);
print out_dir.replace(dir_server,'http://vision1.idav.ucdavis.edu:1000')+out_dir[out_dir.rindex('/'):]+'.html';

    
```
