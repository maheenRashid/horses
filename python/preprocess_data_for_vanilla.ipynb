```python
import visualize;
import util;
import random;
import os;
import numpy as np;
import cv2;
import scipy.misc;

def writeMergedTrainFile(pair_files,out_file_train):
    lines_all=[];
    for file_curr in pair_files:
        lines=util.readLinesFromFile(file_curr);
        lines_all=lines_all+lines;

    random.shuffle(lines_all);
    print len(lines_all),out_file_train;
    util.writeFile(out_file_train,lines_all);



def getMeanImage(im_files,resize_size,disp_idx=1000):
    running_total=np.zeros((resize_size[0],resize_size[1],3));
    for idx_file_curr,file_curr in enumerate(im_files):
        if idx_file_curr%disp_idx==0:
            print idx_file_curr;
        im=cv2.imread(file_curr)
        if im.shape[1]!=resize_size[0] or im.shape[2]!=resize_size[1]:
            im=scipy.misc.imresize(im,resize_size);
        running_total=running_total+im;

    running_total=running_total.astype(np.float);
    mean=running_total/len(im_files);
    print np.min(mean),np.max(mean);
    return mean;

def getSTDImage(im_files,mean,disp_idx):
    running_total=np.zeros(mean.shape);
    resize_size=(mean.shape[0],mean.shape[1])
    for idx_file_curr,file_curr in enumerate(im_files):
        if idx_file_curr%disp_idx==0:
            print idx_file_curr;
        im=cv2.imread(file_curr);
        if im.shape[1]!=resize_size[0] or im.shape[2]!=resize_size[1]:
            im=scipy.misc.imresize(im,resize_size);
#         im=scipy.misc.imresize(im,resize_size);
        std_curr=np.power(mean-im,2)
        running_total=running_total+im;

    running_total=running_total.astype(np.float);
    std=running_total/len(im_files);
    std=np.sqrt(std)
    print np.min(std),np.max(std);
    return std;


pair_file_aflw='/home/SSD3/maheen-data/horse_project/data_check/aflw/pairs.txt'
pair_file_face='/home/SSD3/maheen-data/horse_project/data_check/face/pairs.txt'
# pair_file_horse='/home/SSD3/maheen-data/horse_project/data_check/horse/pairs.txt'
pair_files=[pair_file_aflw,pair_file_face];
out_dir_meta='/home/SSD3/maheen-data/data_face_network';
util.mkdir(out_dir_meta);
dataset_pre='aflw_cvpr';
train_pair_file=os.path.join(out_dir_meta,dataset_pre+'_train.txt');

out_file_mean=os.path.join(out_dir_meta,dataset_pre+'_mean.png');
out_file_std=os.path.join(out_dir_meta,dataset_pre+'_std.png');
resize_size=(224,224);
out_file_mean=os.path.join(out_dir_meta,dataset_pre+'_'+str(resize_size[0])+'_mean.png');
out_file_std=os.path.join(out_dir_meta,dataset_pre+'_'+str(resize_size[0])+'_std.png');

im_files=util.readLinesFromFile(train_pair_file);
im_files=[line_curr[:line_curr.index(' ')] for line_curr in im_files];
# im_files=im_files[:100];
print len(im_files)

mean_im = getMeanImage(im_files,resize_size,disp_idx=1000)
std_im = getSTDImage(im_files,mean_im,disp_idx=1000);

cv2.imwrite(out_file_mean,mean_im);
cv2.imwrite(out_file_std,std_im);
```
